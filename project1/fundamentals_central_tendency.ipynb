```python
import pandas as pd
import numpy as np
```

```python
table=pd.read_excel("C:/Users/ASUS/Downloads/IBM-313 Marks.xlsx")
print(table)
```

```python
#MEAN
x=table['Total']
np.mean(x)
```

```python
#MEDIAN
np.median(x)
```

```python
#MODE
import scipy
from scipy import stats
stats.mode(x)
```

```python
#PERCENTILE
a=np.array([1,2,3,4,5])
p=np.percentile(a,50) #returns the 50th percentile,e.g=median.
print(p)
```

```python
#RANGE
data=[1,3,4,5,6,7,3,9,22,2004,10]
def rangef(data):
    min_val=min(data)
    max_val=max(data)

    return(max_val-min_val)

rangef(data)
```

```python
#QUARTILE
a=np.array([1,2,3,4,5])
q1=np.percentile(a,50) #returns the 25th percentile,e.g=median.
print(q1)
```

```python
a=np.array([1,2,3,4,5])
q2=np.percentile(a,75) #returns the 75th percentile,e.g=median.
print(q2)
```

```python
#INTER QUARTILE
iq=q2-q1
print(iq)
```
