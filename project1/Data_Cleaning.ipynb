```python

df.replace([np.inf],-np.nan,inplace=True)
```

```python
#handling missing values
df.isnull()
df.isnull().sum()

#fill missing values
df.fillna(0, inplace=True)

df.fillna(df.mean(numeric_only=True), inplace=True)

#drop missing values
df.dropna(inplace=True)
```

```python
#Remove duplicate rows
df.duplicated()
df.drop_duplicates(inplace=True)
```

```python
#filtering Data
df[df["Age"] > 30]
df[(df["Age"] > 25) & (df["Salary"] > 50000)]
```

```python
#sorting data
df.sort_values("Salary")
df.sort_values("Salary", ascending=False)
```

```python
#remove negative salaries
df['Salary']=np.where(df['Salary']<0,df['Salary'].mean(),df['Salary'])
```

```python
print(df)
```

```python
salary_mean= df['Salary'].mean()
salary_std= df['Salary'].std()
lower_bound = salary_mean - (3 * salary_std)
upper_bound = salary_mean + (3 * salary_std)
```

```python
df = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]
print(df)
```

CLEANING THE DATA

```python
df.loc[:, 'Age'] = df['Age'].fillna(df['Age'].median()).astype(int)
df.loc[:, 'Salary'] = df['Salary'].fillna(df['Salary'].median()).astype(int)
df.loc[:, 'Performance_Rating'] = df['Performance_Rating'].round(1)
```

```python
columns_to_drop = ['Experience_Years','State','Education','Joining_Date','Projects_Completed','Overtime_Hours']
# Filter columns_to_drop to only include columns that exist in the DataFrame
columns_to_drop_existing = [col for col in columns_to_drop if col in df.columns]
df = df.drop(columns = columns_to_drop_existing)
```
